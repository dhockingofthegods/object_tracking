```python
from ultralytics import YOLO
model = YOLO('yolov8s.pt')

#config
epochs = 30
batch_size = -1
img_size = 640
project_name = 'models/yolo'
name = 'yolov8s_mot17_det'

#train the model
results = model.train(
    data = '/kaggle/input/mot17datasets/MOT17/data.yml',
    epochs = epochs,
    batch = batch_size,
    imgsz = img_size,
    project = project_name,
    name = name
)
```
